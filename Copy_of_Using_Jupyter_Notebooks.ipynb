# Using Jupyter Notebooks



```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import randn,randint,uniform,sample
%matplotlib inline
```

```python
fmri= sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.describe().transpose()
```

```python
sns.relplot(data=fmri,x='timepoint',y='signal',hue ='event',size='event',kind ='line')
plt.title('Relation between timepoint and signal')
plt.show()
```

```python
sns.relplot(y='signal',x='timepoint',kind='line',hue='subject',size='region',data=fmri)
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue= 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.boxplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.boxplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'event')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.lineplot(y='signal',x='timepoint',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=fmri,style='event')
```

```python
z=fmri.query("region == 'frontal'")
sns.lineplot(
    data=fmri.query("region == 'frontal'"),
    x="timepoint", y="signal", hue="event", units="subject",
    estimator=None, lw=1,
)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=z,style='event')
```
